````python
#import de totes les llibreries utilitzades
import os
import zipfile
from genericpath import isfile
import pandas as pd
import re
````

````python
def size_calculator(start_path='.'):
    # Inicialitzem la mida total a 0
    total_size = 0

    # Iterem sobre tots els fitxers i directoris a partir del directori proporcionat (start_path)
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            # Obtenim la ruta completa del fitxer
            fp = os.path.join(dirpath, f)

            # Verifiquem si el fitxer no és un enllaç simbòlic
            if not os.path.islink(fp):
                # Sumem la mida del fitxer a la mida total
                total_size += os.path.getsize(fp)

    # Si el camí proporcionat és un fitxer, actualitzem la mida total amb la mida d'aquest fitxer
    if os.path.isfile(start_path):
        total_size = os.path.getsize(start_path)

    # Convertim la mida total a megabytes (MB)
    total_size = total_size / 1048576

    # Retornem la mida total
    return total_size

# Exemple d'ús amb el directori actual com a directori d'inici
mida_total = size_calculator()
print(f"Mida total del directori o fitxer: {mida_total:.2f} MB")
````

````python
#2.2
#
#
#
def diccionari_original(path):
    i = 0
    finalDic = {}

    # Iterem sobre tots els fitxers en el directori i subdirectoris
    for root, dirs, files in os.walk(path):
        for f in files:
            # Obtenim la ruta completa del fitxer
            fp = os.path.join(root, f)

            # Obrim el fitxer en mode de lectura ('r')
            with open(fp, "r") as text:
                # Llegim totes les línies del fitxer
                linesInAFile = text.readlines()

                # Iterem sobre cada línia del fitxer
                for l in linesInAFile:
                    # Separem la línia en base al caràcter ';'
                    dicKey = l.split(';')

                    # Obtenim la informació del camí del fitxer
                    pathSplitted = fp.split('/')

                    # Creem un diccionari amb les dades necessàries
                    dic =  {
                        'summary': dicKey[0],
                        'city': pathSplitted[4],
                        'state': pathSplitted[3],
                        'date_time': dicKey[1],
                        'shape': f.replace('.txt', ''),
                        'duration': dicKey[2],
                        'text': dicKey[3],
                        'city_latitude': dicKey[4],
                        'city_longitude': dicKey[5]
                    }

                    # Afegim el diccionari al diccionari final amb la clau 'i'
                    finalDic[i] = dic
                    i = i + 1

    return finalDic

# Exemple d'ús amb la ruta del directori
path = "/content/original"
diccionari = diccionari_original(path)

# Mostra el diccionari resultant
print(diccionari)
````

````python
#2.3
#
#
# Converteix el diccionari en un DataFrame
df = pd.DataFrame.from_dict(diccionari, orient='index')

# Guarda el DataFrame com a un fitxer CSV
df.to_csv('fullDF.csv', index_label='IndexLabel')

# Imprimeix les primeres 10 files del DataFrame
print(df.head(10))

# Mostra la longitud del diccionari
print(len(diccionari))
````
